# SCRAPING DATA  FROM A WEBSITE + PANDAS FOR DATA MANIPULATION

```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://en.wikipedia.org/wiki/List_of_companies_of_New_Zealand'
page= requests.get(url)
soup=BeautifulSoup(page.text,'html')
```

```python
soup.find_all('table')[1]
```

```python
<table class="wikitable sortable jquery-tablesorter">
```

```python
table=soup.find_all('table')[1]
```

```python
print(table)
```

```python
NZ_title =table.find_all('th')
```

```python
NZ_title
```

```python
NZ_table_titles = [title.text.strip() for title in NZ_title]
print(NZ_table_titles)
```

# Pandas Data Frame

```python
import pandas as pd
```

```python
NZ_table_titles = [title.text.strip() for title in NZ_title]
df = pd.DataFrame(columns=NZ_table_titles)
print(df)
```

```python
df
```

# Coloumn Data

```python
table.find_all('tr')
```

```python
coloumn_data=table.find_all('tr')
```

```python
for row in coloumn_data:
    row_data = row.find_all('td')
    individual_raw_data = [data.text.strip() for data in row_data]
    print(individual_raw_data)
```

# Extract and append row data

```python
rows = table.find_all('tr')
for row in coloumn_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length= len(df)
    df.loc[length]=individual_row_data
    
   
```
